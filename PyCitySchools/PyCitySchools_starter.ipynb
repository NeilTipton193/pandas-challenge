```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculate summary stats
total_schools = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["Student ID"].unique())
sum_budget = school_data_complete["budget"].sum()
avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()

#calculate percentages
school_data_complete["passing_math"] = school_data_complete["math_score"]>=70
percentage_pass_math = school_data_complete["passing_math"].mean()
school_data_complete["passing_reading"] = school_data_complete["reading_score"]>=70
percentage_pass_reading = school_data_complete["passing_reading"].mean()
school_data_complete["passing_all"] = (school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)
percentage_pass_all = school_data_complete["passing_all"].mean()

#create dataframe
dist_summary = [{"Total Schools":total_schools, 
                 "Total Students":total_students,
                 "Total Budget":sum_budget,
                 "Average Math Score":avg_math, 
                 "Average Reading Score":avg_reading, 
                 "% Passing Math":percentage_pass_math, 
                 "% Passing Reading":percentage_pass_reading, 
                 "% Passing Both":percentage_pass_all }]
dist_summary_df = pd.DataFrame(dist_summary)
dist_summary_df.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#create new dataframe
school_summary_df = school_data_complete.groupby("school_name").mean()
del school_summary_df['Student ID']
del school_summary_df['School ID']
del school_summary_df['size']
del school_summary_df['budget']

#merge with school_data to get missing metrics
new_school_summary_df = pd.merge(school_summary_df, school_data, how="left", on="school_name")
new_school_summary_df

#rename columns
new_school_summary_df = new_school_summary_df.rename(columns={"school_name":"School Name", 
                                       "reading_score":"Average Reading Score", 
                                       "math_score":"Average Math Score",
                                       "passing_math":"% Passing Math",
                                       "passing_reading":"% Passing Reading", 
                                       "passing_all":"% Overall Passing",
                                       "type":"Type", 
                                       "size":"Total Students", 
                                       "budget":"Budget" })

#create per student budget column
new_school_summary_df["Budget per Student"] = new_school_summary_df["Budget"]/new_school_summary_df["Total Students"]

#reformat
new_school_summary_df["Average Reading Score"] = new_school_summary_df["Average Reading Score"].map("{:,.2f}".format)
new_school_summary_df["Average Math Score"] = new_school_summary_df["Average Math Score"].map("{:,.2f}".format)
new_school_summary_df["% Passing Math"] = new_school_summary_df["% Passing Math"].map("{:,.2f}".format)
new_school_summary_df["% Passing Reading"] = new_school_summary_df["% Passing Reading"].map("{:,.2f}".format)
new_school_summary_df["% Overall Passing"] = new_school_summary_df["% Overall Passing"].map("{:,.2f}".format)
new_school_summary_df["Total Students"] = new_school_summary_df["Total Students"].map("{:,.0f}".format)

new_school_summary_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_perf_df = new_school_summary_df.sort_values("% Overall Passing", ascending=False)
top_perf_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score

  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create new db
school_data_spend = school_data_complete
school_data_spend["Budget Per Student"] = school_data_spend["budget"]/school_data_spend["size"]

#Create bins
spend_bins = [0,585,630,645,680]
spend_names = ["<585", "585-630", "630-645", "645-680"]
school_data_spend["Budget Range Per Student"] = pd.cut(school_data_spend["Budget Per Student"],spend_bins,labels=spend_names)

#Setup table
spend_gb = school_data_spend.groupby("Budget Range Per Student").mean()
del spend_gb["Student ID"]
del spend_gb["School ID"]
del spend_gb["size"]
del spend_gb["budget"]
del spend_gb["Budget Per Student"]

spend_gb
```
